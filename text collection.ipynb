```python
import requests as rq
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
from datetime import datetime, date
```

```python
def get_short_news(start=0):
    response = rq.get(
        'https://iss.moex.com/iss/sitenews.json'
        ,params={
            'start':start
        }
    )
    json_data = response.json()['sitenews']
    return pd.DataFrame(columns=json_data['columns'],data=json_data['data'])

get_short_news()
```

```python
def get_full_news_by_id(id):
    response = rq.get(
        f'https://iss.moex.com/iss/sitenews/{id}.json'
    )
    json_data = response.json()['content']
    return pd.DataFrame(columns=json_data['columns'],data=json_data['data'])

get_full_news_by_id(75726)
```

```python
MAX_TEXT_ID = 75797
TEXT_STEP = 50
path_to_text = 'data/texts/moex_title.txt'
for i in tqdm(range(0,MAX_TEXT_ID,TEXT_STEP)):
    short_text = get_short_news(i)
    df = short_text
    if i==0:
        df.to_csv(path_to_text,mode='w',index=False,header=True)
    else:
        df.to_csv(path_to_text,mode='a',index=False,header=False)
```
